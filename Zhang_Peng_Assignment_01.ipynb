# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here

import math
def V(r):
    return (2 / 3) * math.pi * (r ** 3)

even_numbers = []
for number in range(2, 21):
    if number % 2 == 0:
        even_numbers.append(number)

results = []
for r in even_numbers:
    volume = V(r)
    results.append((r, volume))

print("r    V")
for r, volume in results:
    print(f"{r}    {volume:.2f}")
```

['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
import re
import string

# (b) Transform each sentence into a list, where each element corresponds to a word in the sentence.
def sentence_list(sentence):
    return re.findall(r'\w+|[^\w\s]', sentence)

# (d) Remove any punctuation from the sentence and return the sanitized sentence.
def remove_punctuation(sentence):
    return sentence.translate(str.maketrans('', '', string.punctuation))

# (e) Count the number of vowels in the sentence and return the count.
def count_vowels(sentence):
    vowels = "aeiou"
    count = {vowel: 0 for vowel in vowels}
    for char in sentence.lower():
        if char in vowels:
            count[char] += 1
    return count

# (a) Prompt the user to input two sentences.
def main():
    sentence1 = input("Please enter the first sentence: ")
    sentence2 = input("Please enter the second sentence: ")

    # (b) Transform each sentence into a list of words and punctuation
    list1 = sentence_list(sentence1)
    list2 = sentence_list(sentence2)
    print("List of words and punctuations:")
    print(list1 + list2)

    # (c) Compare the two sentences. Print out a list of words that occur in both sentences.
    common = list(set(list1) & set(list2))
    print("Common words in both sentences:")
    print(common)

    # (d) Remove any punctuation from the sentences and print the sanitized sentences.
    sentence1_nopun = remove_punctuation(sentence1)
    sentence2_nopun = remove_punctuation(sentence2)
    print("Sanitized sentences:")
    print("Sentence 1 (sanitized):", sentence1_nopun)
    print("Sentence 2 (sanitized):", sentence2_nopun)

    # (e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences in each of the two sentences separately.
    vowels_count1 = count_vowels(sentence1)
    vowels_count2 = count_vowels(sentence2)
    print("Vowel count in Sentence 1:")
    for vowel, count in vowels_count1.items():
        print(f"{vowel}: {count}")
    print("Vowel count in Sentence 2:")
    for vowel, count in vowels_count2.items():
        print(f"{vowel}: {count}")

if __name__ == "__main__":
    main()

```

# Question 4 (10 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
import math

def calculator(expression):
    try:
        result = eval(expression)
        return result
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."
    except Exception as e:
        return f"Error: Invalid expression. {e}"

def main():
    print("Simple Command-Line Calculator")
    print("Supported operations: +, -, *, /, ** , % , sqrt ")
    print("Type 'exit' to quit the calculator.")

    while True:
        expression = input("\nEnter expression: ").strip()
        if expression.lower() == "exit":
            print("Exiting calculator. Goodbye!")
            break

        if expression.lower().startswith("sqrt"):
            try:
                number = float(expression[4:].strip("()"))
                result = math.sqrt(number)
                print(f"Result: {result}")
            except ValueError:
                print("Error: Invalid number for square root.")
            continue

        result = calculator(expression)
        print(f"Result: {result}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Assignment1_denton_housing.csv')

# (a) Calculate the difference in the number of Occupied Housing Units from year to year
occupied_df = df[df['title_field'] == 'Occupied Housing Units']

# Calculate the difference in occupied housing units for each consecutive year
occupied_df['Occupied Housing Units Diff'] = occupied_df['value'].diff()

# Print the differences in occupied housing units, sorted in ascending order
occupied_diff_sorted = occupied_df['Occupied Housing Units Diff'].sort_values()
print("Occupied Housing Units Differences (sorted):")
print(occupied_diff_sorted)

# (b) Calculate the percentage of vacant and occupied housing units for each year
vacant_df = df[df['title_field'] == 'Vacant Housing Units']
total_df = df[df['title_field'] == 'Total Housing Units']

# Merge the dataframes to have year, vacant, occupied, and total housing units in one dataframe
merged_df = pd.merge(vacant_df[['year', 'value']], occupied_df[['year', 'value']], on='year', suffixes=('_vacant', '_occupied'))
merged_df = pd.merge(merged_df, total_df[['year', 'value']], on='year')

# Calculate the percentages for vacant and occupied housing units
merged_df['Vacant Percentage'] = (merged_df['value_vacant'] / merged_df['value']) * 100
merged_df['Occupied Percentage'] = (merged_df['value_occupied'] / merged_df['value']) * 100

# Print the percentages of vacant and occupied housing units for each year
print("\nPercentage of Vacant and Occupied Housing Units:")
print(merged_df[['year', 'Vacant Percentage', 'Occupied Percentage']])

# (c) Calculate and print the years and values in which the highest number of vacant and occupied housing units were found
# Find the year with the maximum number of vacant housing units
max_vacant_row = vacant_df.loc[vacant_df['value'].idxmax()]
print("\nYear and Value with the highest number of Vacant Housing Units:")
print(f"Year: {max_vacant_row['year']}, Value: {max_vacant_row['value']}")

# Find the year with the maximum number of occupied housing units
max_occupied_row = occupied_df.loc[occupied_df['value'].idxmax()]
print("\nYear and Value with the highest number of Occupied Housing Units:")
print(f"Year: {max_occupied_row['year']}, Value: {max_occupied_row['value']}")

# (d) Calculate the harmonic mean of the total housing units and print it out
# Calculate the harmonic mean of total housing units across all years
total_housing_units = total_df['value']
harmonic_mean = len(total_housing_units) / (1 / total_housing_units).sum()
print("\nHarmonic Mean of Total Housing Units:")
print(harmonic_mean)

```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
import csv
import os

# Function to automatically create an inventory CSV file
def create_inventory_file(filename='inventory.csv'):
    products = []
    for i in range(1, 51):
        product_id = f"{i:03d}"  # Generate IDs in the format 001, 002, 003 ...
        name = f"Product {chr(64 + (i % 26))}"  # Generate names like Product A, Product B, ...
        category = f"Category {((i - 1) // 10) + 1}"  # Every 10 products belong to a different category
        price = round(10 + (i * 1.5), 2)  # Generate prices starting from 10 and increasing
        quantity = (i * 2) % 100  # Generate random quantities
        products.append([product_id, name, category, price, quantity])

    # Write to the CSV file
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['product_id', 'name', 'category', 'price', 'quantity'])  # Write header
        writer.writerows(products)  # Write product data

    print(f"{filename} has been created with 50 products.")

# Product class that contains product information
class Product:
    def __init__(self, product_id, name, category, price, quantity):
        self.product_id = product_id  # Product ID
        self.name = name  # Product name
        self.category = category  # Product category
        self.price = price  # Product price
        self.quantity = quantity  # Product quantity in stock

    def __repr__(self):
        return f"{self.product_id}, {self.name}, {self.category}, ${self.price}, {self.quantity} in stock"

# Product inventory management system that supports basic functionality
class ProductInventorySystem:
    def __init__(self, filename='inventory.csv'):
        self.filename = filename  # CSV filename
        self.products = []  # List to store all products
        self.load_inventory()  # Load existing inventory data

    # Load CSV file data into memory
    def load_inventory(self):
        if os.path.exists(self.filename):
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.DictReader(file)  # Use DictReader to handle column names automatically
                for row in reader:
                    try:
                        # Check if the column names match the expected ones in the CSV file
                        product = Product(
                            row['product_id'],
                            row['name'],
                            row['category'],
                            float(row['price']),
                            int(row['quantity'])
                        )
                        self.products.append(product)
                    except KeyError as e:
                        print(f"Error reading row: Missing expected column {e}")
                        continue
        else:
            print("Inventory file not found, creating new inventory file.")
            self.create_initial_inventory()

    # Create an initial inventory file
    def create_initial_inventory(self):
        # Define initial product data
        products = [
            {"product_id": "001", "name": "Product A", "category": "Category 1", "price": 20.50, "quantity": 10},
            {"product_id": "002", "name": "Product B", "category": "Category 2", "price": 15.00, "quantity": 5},
            {"product_id": "003", "name": "Product C", "category": "Category 1", "price": 30.00, "quantity": 20},
            {"product_id": "004", "name": "Product D", "category": "Category 3", "price": 10.00, "quantity": 50},
            # You can add more products here
        ]

        with open(self.filename, mode='w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=["product_id", "name", "category", "price", "quantity"])
            writer.writeheader()  # Write the header
            for product in products:
                writer.writerow(product)

        print(f"CSV file '{self.filename}' has been created and initial data written.")
        # Reload the data
        self.load_inventory()

    # Save current inventory data to the CSV file
    def save_inventory(self):
        with open(self.filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['product_id', 'name', 'category', 'price', 'quantity'])  # Write header
            for product in self.products:
                writer.writerow([product.product_id, product.name, product.category, product.price, product.quantity])
        print(f"Inventory saved to {self.filename}")

    # Add a new product
    def add_product(self, product_id, name, category, price, quantity):
        # Check if the product already exists
        for product in self.products:
            if product.product_id == product_id:
                print(f"Product with ID {product_id} already exists. Use update to modify.")
                return
        new_product = Product(product_id, name, category, price, quantity)
        self.products.append(new_product)
        print(f"Product {name} added successfully.")
        self.save_inventory()

    # Delete a product
    def delete_product(self, product_id):
        product_to_delete = None
        for product in self.products:
            if product.product_id == product_id:
                product_to_delete = product
                break
        if product_to_delete:
            self.products.remove(product_to_delete)
            print(f"Product {product_to_delete.name} deleted successfully.")
            self.save_inventory()
        else:
            print(f"Product with ID {product_id} not found.")

    # Update product information
    def update_product(self, product_id, name=None, category=None, price=None, quantity=None):
        for product in self.products:
            if product.product_id == product_id:
                if name:
                    product.name = name
                if category:
                    product.category = category
                if price:
                    product.price = price
                if quantity is not None:
                    product.quantity = quantity
                print(f"Product {product_id} updated successfully.")
                self.save_inventory()
                return
        print(f"Product with ID {product_id} not found.")

    # Search for products
    def search_product(self, search_term):
        results = [product for product in self.products if search_term.lower() in product.name.lower() or search_term.lower() in product.category.lower()]
        if results:
            print("\nSearch Results:")
            for product in results:
                print(product)
        else:
            print("No products found matching the search term.")

    # Generate an inventory report
    def generate_report(self):
        print("\nInventory Report:")
        for product in self.products:
            print(f"{product.name} (ID: {product.product_id}) - {product.category}: {product.quantity} in stock, ${product.price} each")

# Main program
def main():
    system = ProductInventorySystem()  # Create an inventory management system instance

    while True:
        print("\nChoose an action:")
        print("1. Add new product")
        print("2. Delete product")
        print("3. Update product")
        print("4. Search product")
        print("5. Generate inventory report")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            # Input new product details
            product_id = input("Enter product ID: ")
            name = input("Enter product name: ")
            category = input("Enter product category: ")
            price = float(input("Enter product price: "))
            quantity = int(input("Enter product quantity: "))
            system.add_product(product_id, name, category, price, quantity)

        elif choice == '2':
            # Delete product
            product_id = input("Enter product ID to delete: ")
            system.delete_product(product_id)

        elif choice == '3':
            # Update product information
            product_id = input("Enter product ID to update: ")
            print("Leave fields empty to keep current values.")
            name = input("Enter new name: ")
            category = input("Enter new category: ")
            price = input("Enter new price: ")
            price = float(price) if price else None
            quantity = input("Enter new quantity: ")
            quantity = int(quantity) if quantity else None
            system.update_product(product_id, name, category, price, quantity)

        elif choice == '4':
            # Search for products
            search_term = input("Enter search term (product name or category): ")
            system.search_product(search_term)

        elif choice == '5':
            # Generate a report
            system.generate_report()

        elif choice == '6':
            # Exit the system
            print("Exiting system.")
            break

        else:
            print("Invalid choice, please try again.")

# Run the program
if __name__ == "__main__":
    create_inventory_file()  # Call the function to create the inventory file before the main program
    main()

```

# Question 7 (20 points)

Create a well desined Python program for a flight booking system. Offer core functions for adding new flights, finding trips, reserving and releasing seating space and creating performance reports. Ensure the system manages edge cases like invalid input, overbooking, and cancellations. Finally, extend the booking system by including advanced features and maintain about 50 flight in the system.

```python
import datetime

# Flight class to represent each flight with its details
class Flight:
    def __init__(self, flight_id, departure_city, arrival_city, departure_time, total_seats, price_per_seat):
        self.flight_id = flight_id  # Flight ID
        self.departure_city = departure_city  # Departure city
        self.arrival_city = arrival_city  # Arrival city
        self.departure_time = departure_time  # Departure time
        self.total_seats = total_seats  # Total number of seats
        self.price_per_seat = price_per_seat  # Price per seat
        self.booked_seats = 0  # Number of seats booked initially set to 0

    def __repr__(self):
        return f"Flight({self.flight_id}, {self.departure_city} -> {self.arrival_city}, {self.departure_time}, {self.total_seats - self.booked_seats} seats available)"

    # Method to reserve seats on a flight
    def reserve_seat(self, num_seats):
        if self.booked_seats + num_seats > self.total_seats:
            print(f"Error: Cannot reserve {num_seats} seats. Only {self.total_seats - self.booked_seats} seats are available.")
            return False
        self.booked_seats += num_seats
        return True

    # Method to release reserved seats (cancel reservation)
    def release_seat(self, num_seats):
        if num_seats > self.booked_seats:
            print(f"Error: Cannot release {num_seats} seats. Only {self.booked_seats} seats are booked.")
            return False
        self.booked_seats -= num_seats
        return True


# Flight booking system to manage all flights and bookings
class FlightBookingSystem:
    def __init__(self):
        self.flights = []  # List to store all flights
        self.MAX_FLIGHTS = 50  # Maximum number of flights allowed in the system

    # Method to add a new flight to the system
    def add_flight(self, flight_id, departure_city, arrival_city, departure_time, total_seats, price_per_seat):
        if len(self.flights) >= self.MAX_FLIGHTS:
            print("Error: Cannot add more flights. Maximum flight capacity reached.")
            return
        flight = Flight(flight_id, departure_city, arrival_city, departure_time, total_seats, price_per_seat)
        self.flights.append(flight)
        print(f"Flight {flight_id} added successfully.")

    # Method to search for flights based on departure city, arrival city, and optionally departure time
    def find_flights(self, departure_city, arrival_city, departure_time=None):
        result = []
        for flight in self.flights:
            if flight.departure_city.lower() == departure_city.lower() and flight.arrival_city.lower() == arrival_city.lower():
                if departure_time and flight.departure_time.date() != departure_time.date():
                    continue
                result.append(flight)
        return result

    # Method to reserve seats on a flight
    def reserve(self, flight_id, num_seats):
        flight = self.get_flight_by_id(flight_id)
        if not flight:
            print(f"Error: Flight with ID {flight_id} not found.")
            return
        return flight.reserve_seat(num_seats)

    # Method to cancel a reservation (release reserved seats)
    def cancel(self, flight_id, num_seats):
        flight = self.get_flight_by_id(flight_id)
        if not flight:
            print(f"Error: Flight with ID {flight_id} not found.")
            return
        return flight.release_seat(num_seats)

    # Method to generate a performance report showing each flight's booking status
    def generate_report(self):
        print("\nFlight Report:")
        for flight in self.flights:
            print(f"{flight.flight_id}: {flight.departure_city} -> {flight.arrival_city}, "
                  f"{flight.booked_seats}/{flight.total_seats} seats booked")

    # Helper method to get a flight by its flight_id
    def get_flight_by_id(self, flight_id):
        for flight in self.flights:
            if flight.flight_id == flight_id:
                return flight
        return None


# Main program to run the booking system
def main():
    system = FlightBookingSystem()  # Initialize the booking system

    while True:
        print("\nChoose an action:")
        print("1. Add new flight")
        print("2. Find flights")
        print("3. Reserve seat")
        print("4. Cancel reservation")
        print("5. Generate report")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            # Input details for a new flight
            flight_id = input("Enter flight ID: ")
            departure_city = input("Enter departure city: ")
            arrival_city = input("Enter arrival city: ")
            departure_time = input("Enter departure time (YYYY-MM-DD HH:MM): ")
            departure_time = datetime.datetime.strptime(departure_time, "%Y-%m-%d %H:%M")  # Parse datetime
            total_seats = int(input("Enter total number of seats: "))
            price_per_seat = float(input("Enter price per seat: "))
            system.add_flight(flight_id, departure_city, arrival_city, departure_time, total_seats, price_per_seat)

        elif choice == '2':
            # Search for flights
            departure_city = input("Enter departure city: ")
            arrival_city = input("Enter arrival city: ")
            departure_time_str = input("Enter departure time (YYYY-MM-DD) or leave empty: ")
            departure_time = None
            if departure_time_str:
                departure_time = datetime.datetime.strptime(departure_time_str, "%Y-%m-%d")  # Parse date
            flights = system.find_flights(departure_city, arrival_city, departure_time)
            if flights:
                for flight in flights:
                    print(flight)  # Display the found flights
            else:
                print("No flights found.")

        elif choice == '3':
            # Reserve seats on a flight
            flight_id = input("Enter flight ID to reserve: ")
            num_seats = int(input("Enter number of seats to reserve: "))
            system.reserve(flight_id, num_seats)

        elif choice == '4':
            # Cancel a reservation (release seats)
            flight_id = input("Enter flight ID to cancel: ")
            num_seats = int(input("Enter number of seats to cancel: "))
            system.cancel(flight_id, num_seats)

        elif choice == '5':
            # Generate and display a performance report
            system.generate_report()

        elif choice == '6':
            # Exit the system
            print("Exiting system.")
            break

        else:
            print("Invalid choice, please try again.")


# Run the program
if __name__ == "__main__":
    main()

```
